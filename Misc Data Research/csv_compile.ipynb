```python
import os
import pandas as pd
import tabula

filepath = 'PDF Bulletins'
```

```python
df_arr = {}

for file in os.listdir(filepath):
    if file.endswith('.csv'):
        meta = file.split(' ')
        week = meta[-1].split('.')[0]
        print(week)

        ## NOTE: Week 5 and Week 26 were tabled wrongly and have some errors
        temp_df = pd.read_csv(os.path.join(filepath, file))

        temp_df = temp_df.rename(columns={
            temp_df.columns[0]: 'Disease',
            temp_df.columns[1]: '2025',
            temp_df.columns[2]: '2024',
            temp_df.columns[3]: 'Median 2020-2024',
            temp_df.columns[4]: 'Cumulative 2025',
            temp_df.columns[5]: 'Cumulative 2024',
            temp_df.columns[6]: 'Cumulative Median 2020-2024'
        })

        temp_df = temp_df.drop([0,1])
        temp_df = temp_df.reset_index()
        temp_df = temp_df.drop(columns = ['index'])
        
        print(list(temp_df))

        df_arr[file] = {
            'data': temp_df,
            'week': week,
            'filename': file
        }
```

```python
df_arr
```

```python
cur_disease = ''
dict_keys = df_arr.keys()

master_df = {
    'Disease': [],
    'Disease Type': [],
    'Year': [],
    'Week': [],
    '2025': [],
    '2024': [],
    'Median 2020-2024': [],
    'Cumulative 2025': [],
    'Cumulative 2024': [],
    'Cumulative Median 2020-2024': []
}

for key in dict_keys:
    df = df_arr[key]

    data = df['data']
    for i, row in data.iterrows():
        
        # NOTE: very very scuffed way to check if it is a new category

        if row.isnull().sum() == 6:
            cur_disease = row[0]

            if 'polyclinic' in str(cur_disease).lower():
                break
            continue

        master_df['Disease'].append(row['Disease'])
        master_df['Disease Type'].append(cur_disease)
        master_df['Year'].append(2025)
        master_df['Week'].append(df['week'])
        master_df['2025'].append(row['2025'])
        master_df['2024'].append(row['2024'])
        master_df['Median 2020-2024'].append(row['Median 2020-2024'])
        master_df['Cumulative 2025'].append(row['Cumulative 2025'])
        master_df['Cumulative 2024'].append(row['Cumulative 2024'])
        master_df['Cumulative Median 2020-2024'].append(row['Cumulative Median 2020-2024'])
    
```

```python
compile_df = pd.DataFrame(master_df)
```
