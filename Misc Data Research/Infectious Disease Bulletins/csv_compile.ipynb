```python
import os
import pandas as pd
import re
import tabula

rootpath = 'PDF Bulletins'
years = [2023,2024,2025]

broken_files = ['2023_week_18-2023.csv']
```

```python
df_arr = {}
broken_csvs = []

for year in years:
    filepath = os.path.join(rootpath, str(year))
    for file in os.listdir(filepath):
        print(file)
        if file.endswith('.csv'):

            if file in broken_files:
                continue

            meta = re.split(r'[\-\_\.\s]+' ,file)
            if meta[-2] == str(year):
                week = meta[-3]
            else:
                week = meta[-2]
            
            print(year)
            print(week)

            ## NOTE: Week 5 and Week 26 were tabled wrongly and have some errors
            temp_df = pd.read_csv(os.path.join(filepath, file))

            temp_df = temp_df.rename(columns={
                temp_df.columns[0]: 'Disease',
                temp_df.columns[1]: f'{year}',
                temp_df.columns[2]: f'{year-1}',
                temp_df.columns[3]: f'Median {year-5}-{year-1}',
                temp_df.columns[4]: f'Cumulative {year}',
                temp_df.columns[5]: f'Cumulative {year-1}',
                temp_df.columns[6]: f'Cumulative Median {year-5}-{year-1}'
            })

            temp_df = temp_df.drop([0,1])
            temp_df = temp_df.reset_index()
            temp_df = temp_df.drop(columns = ['index'])
            
            print(list(temp_df))
            if len(list(temp_df)) != 7:
                broken_csvs.append(file)

            df_arr[file] = {
                'data': temp_df,
                'year': year,
                'week': week,
                'filename': file
            }
```

```python
cur_disease = ''
dict_keys = df_arr.keys()

master_df = {
    'Disease': [],
    'Disease Type': [],
    'Year': [],
    'Week': [],
    'Current Year': [],
    'Previous Year': [],
    'Median Previous 5 Years': [],
    'Cumulative Current Year': [],
    'Cumulative Previous Year': [],
    'Cumulative Median Previous 5 Years': []
}

for key in dict_keys:
    df = df_arr[key]

    data = df['data']
    year = df['year']
    week = df['week']

    print(f"{year}, {week}")

    for i, row in data.iterrows():
        
        # NOTE: very very scuffed way to check if it is a new category

        if row.isnull().sum() == 6:
            cur_disease = row[0]

            if 'polyclinic' in str(cur_disease).lower():
                break
            continue

        master_df['Disease'].append(row['Disease'])
        master_df['Disease Type'].append(cur_disease)
        master_df['Year'].append(int(year))
        master_df['Week'].append(int(week))
        master_df['Current Year'].append(row[f'{year}'])
        master_df['Previous Year'].append(row[f'{year-1}'])
        master_df['Median Previous 5 Years'].append(row[f'Median {year-5}-{year-1}'])
        master_df['Cumulative Current Year'].append(row[f'Cumulative {year}'])
        master_df['Cumulative Previous Year'].append(row[f'Cumulative {year-1}'])
        master_df['Cumulative Median Previous 5 Years'].append(row[f'Cumulative Median {year-5}-{year-1}'])
    
```

```python
compile_df = pd.DataFrame(master_df)
```

```python
compile_df
```

```python
compile_df = compile_df.sort_values(by = ['Year', 'Week'])
compile_df = compile_df.reset_index(drop=True)
```

```python
compile_df.to_csv(f'Weekly Infectious Disease Bulletin_Master_v2.csv')
```
