```python
import os
import pandas as pd
```

```python
mrt_master_df = pd.DataFrame()
```

```python
filepath = 'Data\LTA API\PassengerVolumeByTrainStations'

mrt_dfs = []

for file in os.listdir(filepath):
    if file.startswith('transport_node_train') and file.endswith('.csv'):
        print(file)
        temp_df = pd.read_csv(os.path.join(filepath, file))
        mrt_dfs.append(temp_df)

mrt_master_df = pd.concat(mrt_dfs)
```

```python
mrt_master_df
```

```python
missing_hours = [1,2,3,4]

for year_month in mrt_master_df['YEAR_MONTH'].unique():
    for day_type in mrt_master_df['DAY_TYPE'].unique():
        for pt_type in mrt_master_df['PT_TYPE'].unique():
            for pt_code in mrt_master_df['PT_CODE'].unique():
                for hour in missing_hours:
                    mrt_master_df.loc[len(mrt_master_df)] = [year_month, day_type, hour, pt_type, pt_code, 0, 0]
```

```python
mrt_master_df.to_csv('Data\Master Sets\TrainStationMaster.csv')
```
